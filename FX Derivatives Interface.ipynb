```python
def simul_price_black_scholes(prix_t, sigma, dt, r):
    """Simule le prochain prix d'un actif en utilisant le modèle de Black-Scholes."""
    z = random.gauss(0, 1)  # Utilise random.gauss pour la distribution normale
    prix_t_plus_dt = prix_t * np.exp((r - 0.5 * sigma**2) * dt + sigma * np.sqrt(dt) * z)
    return prix_t_plus_dt
```

```python
class MainWindow(QWidget):
    def __init__(self):
        super().__init__()

        self.prix_actuel = 100.0
        self.volatilite = 0.2  # 20%
        self.increment_temps = 1/252  # Un jour
        self.taux_sans_risque = 0.05  # 5%

        self.prix_label = QLabel(f"Prix : {self.prix_actuel:.2f}")  # Affiche le prix avec 2 décimales
        self.prix_label.setAlignment(Qt.AlignCenter)

        layout = QVBoxLayout()
        layout.addWidget(self.prix_label)
        self.setLayout(layout)

        self.timer = QTimer()
        self.timer.timeout.connect(self.mettre_a_jour_prix)
        self.timer.start(100)  # Mise à jour toutes les 100 millisecondes (ajustez selon vos besoins)

        self.setWindowTitle("Simulation de Prix")

    def mettre_a_jour_prix(self):
        self.prix_actuel = simul_price_black_scholes(self.prix_actuel, self.volatilite, self.increment_temps, self.taux_sans_risque)
        self.prix_label.setText(f"Prix : {self.prix_actuel:.2f}")
```

```python
if __name__ == "__main__":
    app = QApplication(sys.argv)
    main_window = MainWindow()
    main_window.show()
    sys.exit(app.exec())
```
